```python
plt.scatter(df["median_house_value"], df["households"])
plt.xlabel("median_house_value")
plt.ylabel("households")
plt.title("Scatter Plot of Longitude vs. Latitude")
plt.grid(True)  # Add grid lines (optional)
plt.show()
```
